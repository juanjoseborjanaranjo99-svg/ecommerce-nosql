```python
# CELDA 2: cargar los principales CSV de Olist
import pandas as pd

orders = pd.read_csv("olist_orders_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")

print("Orders:", orders.shape)
print("Order items:", order_items.shape)
print("Products:", products.shape)

# Mostrar las primeras filas para comprobar
display(orders.head(2))
display(order_items.head(2))
display(products.head(2))
```

```python
# CELDA 3: unir órdenes + ítems + productos

# Convertir la columna de tiempo a datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

# Merge: items con productos
items_products = order_items.merge(products, on="product_id", how="left")

# Merge: + órdenes (para user_id y timestamp)
df = items_products.merge(
    orders[['order_id', 'customer_id', 'order_purchase_timestamp']],
    on="order_id",
    how="left"
)

print("DataFrame combinado:", df.shape)
display(df.head(3))
```

```python
# CELDA 4: generar eventos JSONL (VIEW, CART, PURCHASE)

import uuid, random, json
from datetime import timedelta

OUTPUT_FILE = "sample_events.jsonl"
TARGET_EVENTS = 2000   # cámbialo a 1000–10000 según lo que quieras
random.seed(42)

def make_session_id(user_id, ts):
    day = ts.strftime("%Y%m%d")
    return f"session_{user_id}_{day}"

def gen_view_times(purchase_ts, n_views):
    start = purchase_ts - timedelta(days=30)
    end = purchase_ts - timedelta(minutes=1)
    return sorted([start + (end - start) * random.random() for _ in range(n_views)])

def gen_cart_time(view_ts, purchase_ts):
    start = view_ts + timedelta(seconds=1)
    if start >= purchase_ts:
        return purchase_ts - timedelta(seconds=1)
    return start + (purchase_ts - start) * random.random()

count = 0
with open(OUTPUT_FILE, "w", encoding="utf-8") as out:
    order_ids = df['order_id'].unique().tolist()
    random.shuffle(order_ids)
    for oid in order_ids:
        items = df[df['order_id'] == oid]
        if items.empty:
            continue
        purchase_ts = pd.to_datetime(items['order_purchase_timestamp'].iloc[0])
        user_id = items['customer_id'].iloc[0]
        session_id = make_session_id(user_id, purchase_ts)

        for _, row in items.iterrows():
            if count >= TARGET_EVENTS: break

            # PURCHASE
            purchase_event = {
                "event_id": str(uuid.uuid4()),
                "event_type": "PURCHASE",
                "user_id": str(user_id),
                "session_id": session_id,
                "product_id": str(row['product_id']),
                "category": str(row['product_category_name']),
                "price": float(row['price']),
                "quantity": 1,
                "order_id": str(oid),
                "event_timestamp": purchase_ts.isoformat(),
                "properties": {}
            }
            out.write(json.dumps(purchase_event) + "\n")
            count += 1

            # VIEWS
            n_views = random.choice([0,1,2,3])
            view_times = gen_view_times(purchase_ts, n_views)
            for vts in view_times:
                if count >= TARGET_EVENTS: break
                view_event = {
                    "event_id": str(uuid.uuid4()),
                    "event_type": "VIEW",
                    "user_id": str(user_id),
                    "session_id": session_id,
                    "product_id": str(row['product_id']),
                    "category": str(row['product_category_name']),
                    "price": float(row['price']),
                    "quantity": 0,
                    "order_id": None,
                    "event_timestamp": pd.to_datetime(vts).isoformat(),
                    "properties": {}
                }
                out.write(json.dumps(view_event) + "\n")
                count += 1

            # CART (opcional)
            if n_views > 0 and random.random() < 0.6 and count < TARGET_EVENTS:
                cart_ts = gen_cart_time(view_times[-1], purchase_ts)
                cart_event = {
                    "event_id": str(uuid.uuid4()),
                    "event_type": "CART",
                    "user_id": str(user_id),
                    "session_id": session_id,
                    "product_id": str(row['product_id']),
                    "category": str(row['product_category_name']),
                    "price": float(row['price']),
                    "quantity": 1,
                    "order_id": None,
                    "event_timestamp": pd.to_datetime(cart_ts).isoformat(),
                    "properties": {}
                }
                out.write(json.dumps(cart_event) + "\n")
                count += 1

        if count >= TARGET_EVENTS: break

print(f" Generados {count} eventos en {OUTPUT_FILE}")
```
